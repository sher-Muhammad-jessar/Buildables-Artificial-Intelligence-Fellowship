# *Design prompts for everyday work tasks* 

Below is a student-style recap that captures what we learned from this guide, organized into clear sections so you can review fast and remember how to apply each idea in your day-to-day work.

## Core idea: make AI part of your workflow

The book’s central message is simple: don’t treat AI as a novelty — treat it as a practical assistant that speeds up the tasks you already do. Instead of asking “what can AI do?” ask “what am I already doing that AI can help with?” This mindset shift reframes AI from an abstract tool into a productivity multiplier for everyday tasks like summarizing, drafting, planning, and tracking.

## The prompt framework in practice

The same basic prompt structure from the first book appears here as the practical backbone: always include persona (who the AI should “be”), context (why this task matters), task (what to produce), format (how to present it), and specificity/constraints (tone, length, budgets, audience). Work flows become faster and more consistent when prompts follow that structure and you iterate on outputs.

## High-value use cases (what to use AI for)

The book lists several everyday actions where AI helps most: summarizing long documents or research papers, drafting emails for different audiences, brainstorming ideas, building tables and trackers, converting meeting transcripts into actionable notes, and creating concise summaries tailored to specific audiences. For each case, the book gives clear prompt patterns and quick iteration examples so you can get usable results on the first or second try.

## Emails: fast drafts and tone control

AI excels at drafting first versions. Use prompts that specify persona (e.g., “I am a gym manager”), context (new schedule), task (inform staff), and tone/format (short, professional, friendly). Then iterate to change audience and tone — the same facts can be rewritten for executives (formal), staff (concise & operational), or customers (friendly & motivating). Always verify factual details before sending.

## Tone & style: teach the AI your voice

Being explicit about tone and giving concrete references produces much better results than vague directions. Instead of “make it funny,” say “witty banter” or “dry wit like a wry comedian.” Provide example phrases or reference voices and iterate until the text fits your intended personality. The more precise your descriptors and examples, the closer the AI will match your desired voice.

## Brainstorming :structured, iterative ideation

Use a persona + context + task + format pattern to prompt brainstorming. Start broad (e.g., one-year launch plan), then iterate with follow-ups that add constraints (global audience, budget limits, cultural differences) or request variants. Treat AI as a partner that surfaces many options quickly; your role is to choose and refine.

## Tables & trackers :fast structure + constraints

Ask the AI to produce structured tables (schedules, project trackers, staff rotas) and give explicit constraints (shift rules, fairness, maximum tasks per person). The AI can output ready-to-copy tables which you paste into Sheets/Excel. Reminder: anonymize sensitive data before using public tools.

## Meeting notes: from messy transcript to clear action items

Upload transcripts and prompt the AI to summarize by speaker, extract key points, and list action items with owners and deadlines. Start with a simple “minutes by speaker” prompt and iterate toward focused formats (action-only lists, thematic grouping). This improves follow-up and accountability.

## Summarizing long documents: chunking & audience tailoring

If a document is too long for a single prompt, chunk it: summarize sections separately, then ask the AI to summarize those summaries into an executive overview. Always include the target audience (executives, small-business owners, students) so the output uses the right language and depth. Use chain-of-density prompting (broad → condensed → one-sentence insight) for clarity.

## Advanced summarization technique — Chain of Density

A recommended 3-step summarization: produce a broad paragraph, then a 3-bullet summary, then a single-sentence insight that captures the main trend. Iteration reduces hallucinations and helps you track which details were removed at each condensation step.

## Responsible use and guardrails

Practical reminders: always protect confidential data, verify facts (human-in-the-loop), watch for hallucinations, and be mindful of bias. Use enterprise or approved tools for sensitive data, anonymize personal info, and validate outputs before acting on them professionally.

## “Always Be Iterating” (ABI): the operating principle

Iteration is the single most important skill. If an output misses the mark, refine the prompt (add constraints, change persona/tone, split tasks). Break large tasks into smaller prompts when needed. Small, specific changes frequently produce much better results than rewriting the whole prompt.

## Quick prompt templates (

Below are compact templates you can reuse immediately.

1. Email draft (staff):
   “You’re a \[persona]. Context: \[one-sentence situation]. Task: Draft a \[tone] email of \~\[length] that \[purpose]. Include \[key facts].”

2. Meeting minutes (action-focused):
   “Summarize this transcript into action items. For each item, list Action, Responsible person, Deadline. Keep entries to one line each.”

3. Weekly schedule (staffing):
   “I have \[N] employees numbered \[1–N]. Make a weekly table with Day, Morning (2 people), Afternoon (2 people). Constraints: no double shifts, fair distribution, no morning after previous-day afternoon.”

4. Document summary (audience-tailored):
   “Summarize this \[X]-page report for \[audience]. Provide: 1) 5-point outline 2) 3 recommended actions that audience can implement tomorrow.”

5. Brainstorming (product launch):
   “You are a \[persona]. Context: launching \[product]. Task: Provide a 6-month timeline + 5 pre-launch buzz ideas. Format: month-by-month bullets + 3 KPIs per stage.”

## Short exercises to practice 

• Draft 3 icebreakers for a teammate who loves basketball and photography.
• Create a 6-month mobile app launch timeline and 3 pre-launch buzz ideas. Then adapt it for a global audience.
• Build a 3-day conference schedule with rule: no speaker more than once per day.
• Summarize a 10-page article into a 10-point bulleted outline, then rewrite it for a high-schooler.
• Convert a meeting transcript into action items with owners and deadlines.

## Final takeaways (student summary)

The guide teaches practical, repeatable patterns: use persona + context + task + format + constraints; iterate; tailor outputs to audience and tone; use chunking for long inputs; and always verify. If you practice a few simple templates (emails, minutes, schedules, summaries) and commit to ABI (Always Be Iterating), you can reliably offload boring, repetitive work to AI and focus on higher-value tasks.
