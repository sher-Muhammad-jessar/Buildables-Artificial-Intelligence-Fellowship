# Reflection Question:

Chain-of-Thought (CoT) style prompting helps convert complex decisions into transparent, auditable steps: it forces you to define criteria, weigh trade-offs, and produce a defensible recommendation with an execution path. Tree-of-Thought (ToT) style prompting multiplies creativity by generating distinct branches (diverse solution concepts) and evaluating them against different constraints. Combining both gives you wide coverage (many candidate ideas) and deep executionability (clear steps for the winner), which is especially useful in fellowship or startup settings where both novelty and reproducibility matter.
